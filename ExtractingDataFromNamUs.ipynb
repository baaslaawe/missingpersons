# Extracting Data from NamUs

The National Missing and Unidentified Persons System (NamUs) is a missing person database funded by the U.S. Department of Justice. NamUs does not provide an API, but does provide a searchable web interface (www.findthemissing.org). 

In this notebook, I want to show how to retrieve and store missing persons data from NamUs using Selenium and store it as JSON. 

```python
import time
import json
from selenium.webdriver.common.by import By
import browsers #from my project
```

Open a new Firefox browser to the NamUs Missing Person database.

```python
browser = browsers.Firefox()
browser.get("https://www.findthemissing.org/en")
```

Try to search for all missing persons in California by selecting California from the drop down menu and clicking submit.

```python
elem = browser.wait_until_clickable("search_Circumstances.StateLKA", timeout=60)
browser.select_option(elem, "california")
submit_btn = browser.find_element_by_name("commit")
submit_btn.click()
#In Ubuntu, I needed an extra click, but Windows I didn't.
#submit_btn.click()
table = browser.wait_until_visible("list", timeout=30)
```

Now the Firefox browser should show a table of search results, but instead of parsing the table, we can parse the Ajax request the page uses to populate the table directly.

```python
browser.get("https://www.findthemissing.org/en/ajax/search_results?page=1&rows=100&sidx=DateLKA&sord=desc&_search=false")
dct = json.loads(browser.find_element_by_css_selector("body").text)
print json.dumps(dct, sort_keys=True, indent=4, separators=(',', ': '))
```

How many pages are in the search results?

```python
int(dct["total"])
```

How many open missing persons cases are there in total for California?

```python
int(dct["records"])
```

There is some interesting information here, but we need to look at the detailed results of each case. 

```python
person = dct["rows"][0]
arr = person["id"].split("_")
browser.get("https://www.findthemissing.org/en/cases/" + arr[0] + "/" + arr[1])
time.sleep(10)
```
