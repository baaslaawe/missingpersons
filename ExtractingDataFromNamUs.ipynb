# Extracting Data from NamUs

<a href="http://www.findthemissing.org">The National Missing and Unidentified Persons System (NamUs)</a> is a missing person database funded by the U.S. Department of Justice. NamUs does not provide an API, but does provide a searchable web interface. 

In this notebook, I want to show how to retrieve missing persons data from NamUs using Selenium. 

```python
import time
import json
from selenium.webdriver.common.by import By
from IPython.display import HTML
import browsers #from my project
```

Open a new Firefox browser to the NamUs Missing Person database.

```python
browser = browsers.Firefox()
browser.get("https://www.findthemissing.org/en")
```

Try to search for all missing persons in California by selecting California from the drop down menu and clicking submit.

```python
elem = browser.wait_until_clickable("search_Circumstances.StateLKA", timeout=60)
browser.select_option(elem, "california")
submit_btn = browser.find_element_by_name("commit")
submit_btn.click()
#In Ubuntu, I needed an extra click, but Windows I didn't.
submit_btn.click()
```

```python
table = browser.wait_until_visible("list", timeout=30)
```

Now the Firefox browser should show a table of search results, but instead of parsing the table, we can parse the Ajax request the page uses to populate the table directly.

```python
browser.get("https://www.findthemissing.org/en/ajax/search_results?page=1&rows=100&sidx=DateLKA&sord=desc&_search=false")
dct = json.loads(browser.find_element_by_css_selector("body").text)
print json.dumps(dct, sort_keys=True, indent=4, separators=(',', ': '))
```

How many pages are in the search results?

```python
int(dct["total"])
```

How many open missing persons cases are there in total for California?

```python
int(dct["records"])
```

There is some interesting information here, but we need to look at the detailed results of each case. 

```python
person = dct["rows"][0]
arr = person["id"].split("_")
browser.get("https://www.findthemissing.org/en/cases/" + arr[0] + "/" + arr[1])
time.sleep(10)
```

```python
print "First name: " + browser.find_element_by_xpath("//div[@id='case_information']/div/table/tbody/tr[2]/td[2]").text
print "Middle name: " + browser.find_element_by_xpath("//div[@id='case_information']/div/table/tbody/tr[3]/td[2]").text
print "Last name: " + browser.find_element_by_xpath("//div[@id='case_information']/div/table/tbody/tr[4]/td[2]").text
```

Next let's determine the missing person's race. It is important to note that in NamUs, race is White, Black/African American, Asian or Pacific Islander, Native American, Other or Unknown and ethnicity is Hispanic/Latino or Other. Therefore a White Hispanic/Latino has race equal to White and ethnicity equal to Hispanic/Latino. A Non-White Hispanic/Latino has race equal to Other and ethnicity equal to Hispanic/Latino.

Look at the person's race. If it is White or Other, then consider if the Ethnicity is relevant.

```python
race = browser.find_element_by_xpath("//div[@id='case_information']/div[2]/table/tbody/tr[3]/td[2]").text

if race == "White" or race == "Other":
    ethnicity = browser.find_element_by_xpath("//div[@id='case_information']/div[2]/table/tbody/tr[4]/td[2]").text
    if ethnicity == "Hispanic/Latino" and race == "White":
        race = "White Hispanic/Latino"
    if ethnicity == "Hispanic/Latino" and race == "Other":
        race = "Non-White Hispanic/Latino"
print "Race: " + race
```

What are the circumstances surrounding this person's disappearance? This information is on the Circumstances page. First click on the Circumstances link from the Case Navigation.

```python
browser.find_element_by_link_text("Circumstances").click()
time.sleep(3)
```

Then print out the city, state and county information.

```python
print "City: " + browser.find_element_by_css_selector("div.column1-unit > table > tbody > tr > td.view_field").text
print "State: " + browser.find_element_by_xpath("//div[@id='circumstances']/div/table/tbody/tr[2]/td[2]").text
print "County: " + browser.find_element_by_xpath("//div[@id='circumstances']/div/table/tbody/tr[4]/td[2]").text
```

The circumstance field may not exist so put it within a try/except block.

```python
try:
    circumstance = browser.find_element_by_id("case_Circumstances").text
except NoSuchElementException:
    circumstance = ""
print "Circumstance: " + circumstance
```

Gather more physical characteristics on the Physical/Medical page.

```python
browser.find_element_by_link_text("Physical / Medical").click()
time.sleep(3)
```

Hair color is relatively straightforward.

```python
hair_color = browser.find_element_by_xpath("//div[@id='physical_characteristics']/div/table/tbody/tr/td[3]").text
print "Hair color: " + hair_color
```

Eye color is more tricky because the person may have two different eye colors. 

```python
left_eye_color = browser.find_element_by_xpath("//div[@id='physical_characteristics']/div/table/tbody/tr[5]/td[3]").text		
right_eye_color = browser.find_element_by_xpath("//div[@id='physical_characteristics']/div/table/tbody/tr[6]/td[3]").text
if left_eye_color == right_eye_color:
    eye_color = left_eye_color
else:
    eye_color = "Multicolor"
print "Eye color: " + eye_color
```
